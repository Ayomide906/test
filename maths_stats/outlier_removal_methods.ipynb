Outlier removal using zscore and std

```python
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(10,6)
```

```python
df3=pd.read_csv('heights.csv')
df3.sample(5)
```

```python
plt.hist(df3.height,bins=20,rwidth=0.6)
plt.xlabel('Height (inches)')
plt.ylabel('count')
plt.show()
```

```python
upper_limit=df3.height.mean()+3*df3.height.std()
lower_limit=df3.height.mean()-3*df3.height.std()
```

```python
df3[(df3.height<lower_limit) | (df3.height> upper_limit)]   #even legitimate values can be an outlier
```

```python
df3_no_outliers=df3[(df3.height>lower_limit) & (df3.height< upper_limit)]
df3_no_outliers.shape
```

```python
#using z score
df3['z_score']=(df3.height-df3.height.mean())/df3.height.std()
df3.head()
```

```python
df3[(df3.z_score>3) |(df3.z_score<-3)] #these are the outliers acc to z score
```
