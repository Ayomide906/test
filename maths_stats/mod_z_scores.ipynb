```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("movie_revenues.csv")
df.head()
```

```python
df["revenue_mln"]=df["revenue"].apply(lambda x:x/1000000)
```

```python
_, mean,std,*_=df.revenue_mln.describe()   #to unpack count  and the rest
mean,std
```

```python
def get_z_score(value,mean,std):
    return (value-mean)/std
```

```python
df['z_score']=df.revenue_mln.apply(lambda x:get_z_score(x,mean,std))
df.z_score.head()
```

```python
df.head(3)
```

```python
#get mad which is better than std 
def get_mad(s):
    median =np.median(s)
    diff=abs(s-median)
    return np.median(diff)
```

```python
mad=get_mad(df.revenue_mln)
median=np.median(df.revenue_mln)
mad,median.round(2)
```

```python
def get_mod_z_score(value,median,mad):
    z_score=0.6745*(value-median)/mad
    return z_score
```

```python
df["modi_z_score"]=df.revenue_mln.apply(lambda x:get_mod_z_score(x,median,mad))
df.head(3)
```

```python
df=df[df.modi_z_score < 3.5]
df.info()
```
