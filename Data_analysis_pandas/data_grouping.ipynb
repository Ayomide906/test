```python
def grouper(df,idx,col):          #set specifics groups you want
    if 80<=df[col].loc[idx]<=90:
        return "80-90"
    elif 50<= df[col].loc[idx]<=60:
        return "50-60"
    else:
        return "others"
```

```python
g=df.groupby(lambda x:grouper(df,x,"temperature")) #uses the custom grouping
g
```

```python
for temp,data in g:
    print(temp)
    print("\n")
    print(data)
```
