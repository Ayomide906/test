```python
df_movies=pd.read_excel("movies_db.xlsx","movies")
df_movies.head(4)
```

```python
df_financials=pd.read_excel("movies_db.xlsx","financials")
df_financials.head(5)
```

```python
def standardize_currency(curr):
    if curr=="$$" or curr=="Dollars":
        curr="USD"
    return curr
df_financials=pd.read_excel("movies_db.xlsx","financials",converters={"currency":standardize_currency})
df_financials.head(5)
```

## <span style="color:blue">To merge two Excel sheets</span>

```python
df_merged=pd.merge(df_movies,df_financials, on="movie_id")
df_merged.head(5)
```

## <span style="color:blue">To export to Excel file</span>


```python
df_merged.to_excel("movies_merged.xlsx",sheet_name="sheet1",index=False)
```

```python
df_stocks=pd.DataFrame({
    "tickers":["GOOGL","WMT","MSFT"],
    "price":[845,65,64],
    "pe":[27.82, 4.61, 2.12],
    "eps":[30.37,15.78,35]
})
df_stocks
```

```python
df_weather=pd.DataFrame({
    'day':['1/1/2025','1/2/2025','1/3/2025'],
    'temperature':[32,35,28],
    'event':['Rain','Sunny','Snow']
})
df_weather
```

## <span style="color:blue">To merge two DataFrames into an excel format</span>

```python
with pd.ExcelWriter("stocks_weather.xlsx") as writer:
    df_stocks.to_excel(writer,sheet_name='stocks')
    df_weather.to_excel(writer,sheet_name='weather')
```
