```python
import pandas as pd
df_sales =pd.read_excel("8_matplotlib_seaborn/linechart.xlsx")
df_sales.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color='blue',label="Fridge")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],color='green',label="Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],color='orange',label="washing machine")
plt.title("Product Sales")
plt.ylabel("Revenue (min $)")
plt.xlabel("Financial Quarter")
plt.legend() #this will create a fine label
plt.show()
```

```python
total_sales=df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales.index
```

<h2 style="color:#001f3f; font-size:32px;">Pie Chart</h2>

```python
plt.pie(total_sales,labels=total_sales.index,colors=["green","blue","orange"],autopct="%1.1f%%",explode=(0.1,0,0),shadow=True,
       startangle=140)    #the labels is passed as a list
plt.show()
```

<h2 style="color:#001f3f, font-size:32px;">Bar Chart</h2>

```python
df_sales.plot(kind="bar",x="Quarter")
plt.xticks(rotation=45)
plt.title("sale analysis quarterly")
```

<h2 style="color:black; font:32px">Histogram</h2>

```python
df_score=pd.read_excel("8_matplotlib_seaborn/histograms.xlsx")
df_score.head()
```

```python
import seaborn as sns
sns.histplot(df_score["Exam_Score"],kde=True)  #kde=True will plot the population density
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.title("Exam Scores Distribution")
plt.show()
```

<h2 style="color:black; font-value:32px">Scatter Plot with many variables</h2>

```python
df_long = df_sales.melt(id_vars="Quarter", 
                        value_vars=["Fridge", "Dishwasher", "Washing Machine"],
                        var_name="Appliance", 
                        value_name="Sales")
sns.scatterplot(data=df_long, x="Quarter", y="Sales", hue="Appliance")

# Optional: improve layout
plt.title("Sales by Appliance Over Quarters")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
