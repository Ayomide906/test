              #statistics of imdb_rating using python

```python
import csv
def calculate_rating_stats(data,industry=None):
    rating=[]
    for row in data:
        if (row[1]==industry or not industry) and row[3]!='NULL':
            rating.append(float(row[3]))
    max_rating=max(rating)
    min_rating=min(rating)
    avg_rating=round(sum(rating)/len(rating),2)
    return max_rating, min_rating, avg_rating
with open("movies.csv") as f:
    data=list(csv.reader(f))
    header=data[0]
    data=data[1:]
max_rating,min_rating,avg_rating=calculate_rating_stats(data)
print("All records: max_rating=",max_rating,",min_rating=",min_rating,", avg_rating=",avg_rating)
max_rating,min_rating,avg_rating=calculate_rating_stats(data,"Hollywood")
print("Hollywood: max_rating=",max_rating,",min_rating=",min_rating,", avg_rating=",avg_rating)
max_rating,min_rating,avg_rating=calculate_rating_stats(data,"Bollywood")
print("Bollywood: max_rating=",max_rating,",min_rating=",min_rating,", avg_rating=",avg_rating)
```
